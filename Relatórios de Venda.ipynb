## Filtro de Venda por Estados 

```python
import pandas as pd
import openpyxl
from openpyxl.styles import Border, Side


df = pd.DataFrame(Relatorio_geral, columns=['UF', 'Pet Shop', 'ID', 'Compra Maio 2023 (kg)'])
estados_filtrados = ['DF', 'ES', 'GO', 'MG', 'MS', 'MT', 'RJ']
df[df['UF'].isin(estados_filtrados)].to_excel('Estados_Filtrados.xlsx', sheet_name='vendas5', index=False, startrow=3)
book = openpyxl.load_workbook('Estados_Filtrados.xlsx')
sheet = book['vendas5']
thin_border = Border(left=Side(style='thin'), 
                     right=Side(style='thin'), 
                     top=Side(style='thin'), 
                     bottom=Side(style='thin'))
for row in sheet.iter_rows(min_row=4, max_row=sheet.max_row, min_col=1, max_col=sheet.max_column):
    for cell in row:
        cell.border = thin_border
sheet.insert_rows(0, 2)
sheet['A1'] = 'Ração Premium - Distribuição - Maio'
sheet['A2'] = 'Maio - 23 - ' + ' '.join(estados_filtrados)
sheet.merge_cells('A1:G1')
sheet.merge_cells('A2:G2')
ultima_linha = sheet.max_row
sheet.cell(row=ultima_linha + 2, column=1, value='Assinado')
book.save('Relatório Venda - Ração Premium.xlsx')

print('Tarefa concluída com sucesso.')
```
